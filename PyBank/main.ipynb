```python
# Import the pathlib and csv library
from pathlib import Path
import csv

# Set the file path
csvpath = Path('../PyBank/budget_data.csv')
```

```python
# Initialize variables
Profit_losses=[]
Date=[]

# Initialize line_num variable
line_num = 0

# Open the input path as a file object
with open(csvpath, 'r') as csvfile:

    # Print the datatype of the file object
    print(type(csvfile))

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=',')
    # Print the datatype of the csvreader
    print(type(csvreader))

    # Go to the next row from the start of the file
    # (which is often the first row/header) and iterate line_num by 1
    header = next(csvreader)
    line_num += 1
    # Print the header
    print(f"{header} <---- HEADER")

    # Read each row of data after the header
    for row in csvreader:
        # Print the row
        print(row)
        # Set profit/loss variable equal to the value in the 2nd column of each row
        profitlosses = int(row[1])
        # Append the row profit/loss value to the list of Profit_losses
        Profit_losses.append(profitlosses)
        #Set month_year variable equal to the value in the 1st column of each row
        month_year = row[0]
        # Append the row date value to the date
        Date.append(month_year)

    # Initialize metric variable
    Total_Number_of_Months=0
    Net_Total_PL=0
    Avg_Change_PL=0
    Max_PL=0
    Max_PL_Period=0
    Min_PL=0
    Min_PL_Period=0
```

```python
# 1. The total number of months included in the dataset
Total_Number_of_Months=len(Date)
```

```python
# 2. The net total amount of Profit/Losses over the entire period
Net_Total_PL=sum(Profit_losses)
```

```python
# 3. The average of the changes in Profit/Losses over the entire period
change=[]
previous_value=0
difference=0
change_average=0
max_change_period_profit=0
max_change_period=0
for i in Profit_losses:
    if previous_value==0:
            previous_value=i
    else:
            difference=i-previous_value
            change.append(difference)
            previous_value=i
Avg_Change_PL=round(sum(change)/len(change),2)
```

```python
# 4. The greatest increase in profits (date and amount) over the entire period
change=[]
previous_value=0
difference=0
change_average=0
max_change_period_profit=0
max_change_period=0
for i in Profit_losses:
    if previous_value==0:
            previous_value=i
    else:
            difference=i-previous_value
            change.append(difference)
            previous_value=i

Max_PL_Period=Date[change.index(max(change))+1]
Max_PL=Profit_losses[change.index(max(change))+1]-Profit_losses[change.index(max(change))]
```

```python
#5. The greatest decrease in losses (date and amount) over the entire period.
change=[]
previous_value=0
difference=0
change_average=0
max_change_period_profit=0
max_change_period=0
for i in Profit_losses:
    if previous_value==0:
            previous_value=i
    else:
            difference=i-previous_value
            change.append(difference)
            previous_value=i

Min_PL_Period=Date[change.index(min(change))+1]
Min_PL=Profit_losses[change.index(min(change))+1]-Profit_losses[change.index(min(change))]
```

```python
#6. Print the analysis results in the terminal
print("Financial Analysis")
print("----------------------------")
print(f"Total Months: {Total_Number_of_Months}")
print(f"Total: ${Net_Total_PL}")
print(f"  Average  Change: ${Avg_Change_PL}")
print(f"Greatest Increase in Profits: {Max_PL_Period} (${Max_PL})")
print(f"  Greatest Decrease in Profits: {Min_PL_Period} (${Min_PL})")
```

```python
# Output to text file
with open("Output.txt", "w") as text_file:
    print("Financial Analysis", file=text_file)
    print("----------------------------", file=text_file)
    print(f"Total Months: {Total_Number_of_Months}", file=text_file)
    print(f"Total: ${Net_Total_PL}", file=text_file)
    print(f"  Average  Change: ${Avg_Change_PL}", file=text_file)
    print(f"Greatest Increase in Profits: {Max_PL_Period} (${Max_PL})", file=text_file)
    print(f"  Greatest Decrease in Profits: {Min_PL_Period} (${Min_PL})", file=text_file)
```
